```python
import pandas as pd
import numpy as np
import io
import cv2

data = pd.read_csv('train.csv')
data_prepared_t = data.drop(['Ticket','Cabin','Embarked','Name'], axis =1)
data_prepared_t = data_prepared_t.fillna(value = np.mean(data['Age']))
data_prepared_t['Male'] = data_prepared_t['Sex'] == 'male'
data_prepared_t = data_prepared_t.drop('Sex', axis=1)
import matplotlib.pyplot as plt
```

```python
max_Age = max(data_prepared_t['Age'])
```

```python
#plt.show()

#fig.canvas.draw()
#data = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
#data = data.reshape(fig.canvas.get_width_height()[::-1] + (3,))

# define a function which returns an image as numpy array from figure
def get_img_from_fig(fig, dpi=180):
    buf = io.BytesIO()
    fig.savefig(buf, format="png", dpi=dpi)
    buf.seek(0)
    img_arr = np.frombuffer(buf.getvalue(), dtype=np.uint8)
    buf.close()
    img = cv2.imdecode(img_arr, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    return img
```

```python
d1 = 65
duration = d1 + 3
fps = 1
def make_frame(t):
    if t<= d1:
        data_prepared = data_prepared_t[data_prepared_t['Age']<= (t+1)/d1*max_Age]
        colors = data_prepared['SibSp'].values + data_prepared['Parch'].values 
        colors = np.round(colors/(max(colors)-min(colors))*32+95)
        area = (max(data_prepared['Pclass']+1)-data_prepared['Pclass'].values)
        area_t = data_prepared['Survived'].values + (1-data_prepared['Male'].values)
        area1 = np.ma.masked_where(area_t == 2 , area)
        area2 = np.ma.masked_where(area_t == 0, area)
        area3 = np.ma.masked_where(area_t == 1, area)
        fig, ax = plt.subplots()
        min_note= 30
        max_note= 80
        note_range = max_note-min_note
        time_range = 64
        x = np.log(data_prepared['Age'].values+np.mean(data_prepared['Age']))
        x = np.round((x-np.min(x))/(np.max(x)-np.min(x))*time_range)
        y = np.log(data_prepared['Fare'].values+ np.mean(data_prepared['Fare']))
        y = np.round((y-np.min(y))/(np.max(y)-np.min(y))*note_range + min_note)
        plt.scatter(x ,y ,s = area1*10, c = colors , marker = 's', cmap ='Greys')
        plt.scatter(x ,y ,s = area2*20, c = colors , marker = 'o', cmap = 'OrRd') #area multipication is just for image beauty
        plt.scatter(x ,y ,s = area3*5, c = colors , marker = 'p', cmap = 'pink')
        clb = plt.colorbar()
        ax.set_facecolor('khaki')
        plt.xlabel("Time-Axis")
        plt.ylabel("Note-Axis")
        clb.ax.tick_params(labelsize=8) 
        clb.ax.set_title('Velocity',fontsize=8)
        plt.title("Titanic Data Sonification")
        plt.yticks(y,[])
        plt.xticks(x,[])
        plot_img_np = get_img_from_fig(fig)
    else:
        data_prepared = data_prepared_t[data_prepared_t['Age']<= t/d1*max_Age]
        colors = data_prepared['SibSp'].values + data_prepared['Parch'].values 
        colors = np.round(colors/(max(colors)-min(colors))*32+95)
        area = (max(data_prepared['Pclass']+1)-data_prepared['Pclass'].values)
        area_t = data_prepared['Survived'].values + (1-data_prepared['Male'].values)
        area1 = np.ma.masked_where(area_t == 2 , area)
        area2 = np.ma.masked_where(area_t == 0, area)
        area3 = np.ma.masked_where(area_t == 1, area)
        fig, ax = plt.subplots()
        min_note= 30
        max_note= 80
        note_range = max_note-min_note
        time_range = 64
        x = np.log(data_prepared['Age'].values+np.mean(data_prepared['Age']))
        x = np.round((x-np.min(x))/(np.max(x)-np.min(x))*time_range)
        y = np.log(data_prepared['Fare'].values+ np.mean(data_prepared['Fare']))
        y = np.round((y-np.min(y))/(np.max(y)-np.min(y))*note_range + min_note)
        plt.scatter(x ,y ,s = area1*10, c = colors , marker = 's', cmap ='Greys')
        plt.scatter(x ,y ,s = area2*20, c = colors , marker = 'o', cmap = 'OrRd') #area multipication is just for image beauty
        plt.scatter(x ,y ,s = area3*5, c = colors , marker = 'p', cmap = 'pink')
        clb = plt.colorbar()
        ax.set_facecolor('khaki')
        plt.xlabel("Time-Axis")
        plt.ylabel("Note-Axis")
        clb.ax.tick_params(labelsize=8) 
        clb.ax.set_title('Velocity',fontsize=8)
        plt.title("Titanic Data Sonification")
        plt.yticks(y,[])
        plt.xticks(x,[])
        plot_img_np = get_img_from_fig(fig)
    return plot_img_np # returns a 8-bit RGB array

audio_background2 = mpy.AudioFileClip('Titanic_Dataset.mp3')
clip2 = mpy.VideoClip(make_frame, duration= duration)
final_clip2 = clip2.set_audio(audio_background2)
#final_clip2.write_videofile("s2.mp4",fps= fps)
```
